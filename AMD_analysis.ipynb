```python
del stock_data["Dividends"]
del stock_data["Stock Splits"]
```

```python
stock_data["Tomorrow"] = stock_data["Close"].shift(-1)
```

```python
stock_data
```

```python
stock_data["Target"] = (stock_data["Tomorrow"] > stock_data["Close"]).astype(int)
```

```python
stock_data
```

```python
stock_data = stock_data.loc["2021-01-01":].copy()
```

```python
stock_data
```

```python
from sklearn.ensemble import RandomForestClassifier
#increase n_estimators for higher accuracy
#min_samples_split protects against overfitting(higher=less accurate and less overfitting)
model = RandomForestClassifier(n_estimators=1000, min_samples_split = 100, random_state=1)

train = stock_data.iloc[:-100]
test = stock_data.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds= model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined= pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=250, step=100):
    all_predictions = []

    for i in range(start, len(data)-step, step):
        train = data.iloc[:i].copy()
        test = data.iloc[i:i+step].copy()
        if test.empty:
            continue
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    if all_predictions:
        return pd.concat(all_predictions)
    else:
        raise ValueError("No predictions generated during backtest")

predictors = ["Close", "Volume", "Open", "High", "Low"]

# Perform backtesting
try:
    predictions = backtest(stock_data, model, predictors)
    print("Backtesting completed successfully.")
except ValueError as e:
    print(f"Backtesting failed: {e}")
```

```python
predictions = backtest(stock_data, model, predictors)
```

```python
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = stock_data.rolling(horizon).mean()

    ratio_column= f"close_Ratio_{horizon}"
    stock_data [ratio_column] = stock_data ["Close"] / rolling_averages ["Close"]

    trend_column = f"Trend_{horizon}"
    stock_data [trend_column] = stock_data.shift(1).rolling(horizon).sum() ["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
stock_data
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.6] = 1
    preds[preds <.6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(stock_data, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
